```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loaded = np.loadtxt('../delivery_fleet_driver.csv', skiprows=1)
```

```python
data = loaded[:,1:]
```

```python
elbow = []
for i in range(1, 5):
    kmeans = KMeans(i)
    kmeans.fit(data)
    elbow.append(kmeans.inertia_)
```

```python
plt.plot(range(1,5), elbow,)
plt.title('metodo elbow')
plt.show()
```

```python
kmeans = KMeans(2)
clt = kmeans.fit_predict(data)
```

```python
plt.scatter(data[clt == 0, 0], data[clt == 0, 1], c='b', label='cluster 1')
plt.scatter(data[clt == 1, 0], data[clt == 1, 1], c='g', label='cluster 2')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='r', label='centroides')
plt.xlabel('distancia')
plt.ylabel('velocidade')
plt.legend()
plt.show()
```
