```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pydataset import data
```

```python
iris = data('iris')
```

```python
plt.scatter(iris['Sepal.Length'], iris['Sepal.Width'], s=20*iris['Petal.Length'])
```

```python
def color(x):
    if x == 'setosa':
        return 'r'
    elif x =='versicolor':
        return 'b'
    else:
        return 'k'
iris['SpeciesNumber'] = iris['Species'].apply(color)
```

```python
plt.scatter(iris['Sepal.Length'], iris['Sepal.Width'], s=20*iris['Petal.Length'], c=iris['SpeciesNumber'])
```
