```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../iris-dataset/iris.data.txt')
```

```python
x = df.iloc[:, [0,1,2,3]].values
```

```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('Metodo do cotovelo')
plt.xlabel('Numero de clusters')
plt.ylabel('soma dos quadrados dos cluters')
plt.show()
```

```python
kmeans = KMeans(3)
y = kmeans.fit_predict(x)
plt.scatter(x[y == 0, 0], x[y == 0, 1], c='r', label='Iris-Setosa')
plt.scatter(x[y == 1, 0], x[y == 1, 1], c='b', label='Iris-Versicolo')
plt.scatter(x[y == 2, 0], x[y == 2, 1], c='g', label='Iris-Virginica')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='y', label='centroides')
plt.show()
```
