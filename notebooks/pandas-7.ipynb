```python
import pandas as pd
import numpy as np
```

```python
dados = {
    'nomes': ['joão', 'maria', 'josé', np.nan, 'pedro', 'judas', 'tiago'],
    'sexo': ['m', 'f', 'm', np.nan, 'm', 'm', np.nan],
    'idade': [14, 13, np.nan, np.nan, 15, 13, 14],
}
```

```python
df = pd.DataFrame(dados)
df
```

```python
df.dropna()
```

```python
df.dropna(how='all')
```

```python
df.dropna(how='all', axis=1)
```

```python
df.dropna(thresh=3)
```

```python
df.fillna(8, inplace=True)
df
```
