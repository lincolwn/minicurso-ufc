```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../primary-results.csv')
df.head()
```

```python
pd.pivot_table(df, index=['state', 'party', 'candidate'], values=['votes'], aggfunc={'votes': np.sum})
```

```python
df['rank'] = df.groupby(['county', 'party'])['votes'].rank(ascending=False)
```
